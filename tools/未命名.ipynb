```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
 
```

```python

def read_tensorboard_data(tensorboard_path, val_name):
    """读取tensorboard数据，
    tensorboard_path是tensorboard数据地址val_name是需要读取的变量名称"""
    ea = event_accumulator.EventAccumulator(tensorboard_path)
    ea.Reload()
    print(ea.scalars.Keys())
    val = ea.scalars.Items(val_name)
    return val
 
def draw_plt(val, val_name):
    """将数据绘制成曲线图，val是数据，val_name是变量名称"""
    plt.figure()
    plt.plot([i.step for i in val], [j.value for j in val], label=val_name)
    """横坐标是step，迭代次数
    纵坐标是变量值"""
    plt.xlabel('step')
    plt.ylabel(val_name)
    plt.show()
 
```

```python
tensorboard_path = '/media/bihuasky/Data/study/school/thesis/ch/nanocenternet_all_voc/logs/Train_loss_loss_train/events.out.tfevents.1614342391.6035ff82c4129e28e385f433.5278.1'
val_name = 'Train_loss/loss'
val = read_tensorboard_data(tensorboard_path, val_name)
draw_plt(val, val_name)
```

```python
plt.figure(figsize=(20,10))
plt.grid()
plt.plot([i.step for i in val], [j.value for j in val], marker='',markevery=1000,label='w',)
plt.plot([i.step for i in val2], [j.value for j in val2], '-',label='w/o')
"""横坐标是step，迭代次数
纵坐标是变量值"""
plt.xlabel('step',fontsize=20)
plt.ylabel('loss',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
